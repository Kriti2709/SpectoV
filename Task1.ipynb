Clustering is a fundamental task in unsupervised machine learning used to discover natural groupings in data. Traditional clustering algorithms like K-Means are efficient for linearly separable and spherical data but fail to identify arbitrarily shaped clusters or noise. This task aims to compare K-Means and DBSCAN clustering algorithms on datasets with complex, non-linear shapes (e.g., moon-shaped data) to analyze their effectiveness in handling non-convex clusters and noise. The goal is to identify which algorithm is better suited for real-world datasets that are irregular in shape and contain outliers

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
#Generating moon shaped dataset
X, y = make_moons(n_samples=300, noise=0.07, random_state=42)

#Plotting the dataset
plt.scatter(X[:, 0], X[:, 1])
plt.title("Moon-Shaped Data")
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.title("K-Means Clustering")
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='plasma')
plt.title("DBSCAN Clustering")
plt.show()
```

Here we know that K-means works well with spherical and well separated data the number of clusters are predefined, but fails when the data is irregular in shape or noisy which is evident in the above codes.
whereas DBSCAN can detect clusters in any shape and can also identify noise and outliners, this doesnt need the number of clusters in advance.
